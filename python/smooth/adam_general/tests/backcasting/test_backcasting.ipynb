```python
#### Global level and ETS(A,N,N) ####
np.random.seed(33)  # For reproducibility
n_points = 100
time_series = np.random.normal(100, 10, n_points)  # Random errors
ts_df = pd.DataFrame({'value': time_series}, index=pd.date_range(start='2023-01-01', periods=n_points, freq='ME'))
```

```python

# we have initial equal to optimal by default
model = 'ANN'
lags = [12]
adam = ADAM(model, lags, initial='optimal')
adam.fit(ts_df)
#print(adam.adam_created)
fc = adam.predict(h = 12)
fc
```

```python
initial = 'optimal'
model = 'ANN'
lags = [12]
adam = ADAM(model, lags, initial = initial)
adam.fit(ts_df)
#print(adam.adam_created)
fc = adam.predict(h = 12)
fc
```

```python
initial = 'backcasting'
model = 'ANN'
lags = [12]
adam = ADAM(model, lags, initial = initial, n_iterations=1)
adam.fit(ts_df)
#print(adam.adam_created)
fc = adam.predict(h = 12)
fc
```

```python
initial = 'backcasting'
model = 'ANN'
lags = [12]
adam = ADAM(model, lags, initial = initial, n_iterations=2)
adam.fit(ts_df)
#print(adam.adam_created)
fc = adam.predict(h = 12)
fc
```
